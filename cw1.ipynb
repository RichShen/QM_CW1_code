```python

from statsmodels.formula.api import ols

#fit simple linear regression model
model = ols('obesity_ratio_diff_10yr ~ clean_air', data=df).fit()

#view model summary
print(model.summary())
```

```python
import statsmodels.api as sm
#define figure size
fig = plt.figure(figsize=(12,8))

#produce regression plots
fig = sm.graphics.plot_regress_exog(model, 'clean_air', fig=fig)
```

```python

from statsmodels.stats.outliers_influence import variance_inflation_factor 
from statsmodels.tools.tools import add_constant

def drop_column_using_vif_(df, thresh=5):

    while True:
      
        df_with_const = add_constant(df)

        vif_df = pd.Series([variance_inflation_factor(df_with_const.values, i) 
               for i in range(df_with_const.shape[1])], name= "VIF",
              index=df_with_const.columns).to_frame()
       
        vif_df = vif_df.drop('const')        
        
        if vif_df.VIF.max() > thresh:
            index_to_drop = vif_df.index[vif_df.VIF == vif_df.VIF.max()].tolist()[0]
            print('Dropping: {}'.format(index_to_drop))
            df = df.drop(columns = index_to_drop)
        else:
            break

    return df
```

```python
df2 = df[['clean_air','clean_environ','health_training','school_awareness','media_awareness','sub_counselling','obesity_ratio_diff_10yr']]
df2_new = drop_column_using_vif_(df2.drop('obesity_ratio_diff_10yr', axis=1))
```
